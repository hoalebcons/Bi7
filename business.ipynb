```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from tqdm import tqdm
```

```python
df = pd.read_csv('Data\\totalsale.csv')
df
```

```python
pr = pd.read_csv('Data\\Dim_product.csv')
pr
```

```python
prt = pr[["product_id","current_cost"]]
```

```python
df = pd.merge(df, prt, on='product_id',how='left')
df
```

```python
df["Total_Sale"] = df["quantity_sold"] * df["unit_price"]
df
```

```python
df["Profit"] = (df["unit_price"] - df["current_cost"]) *  df["quantity_sold"]
df
```

```python
df.to_csv("totalsale.csv")
```
